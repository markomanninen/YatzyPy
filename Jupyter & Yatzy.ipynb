### Show Yatzy categories with scores and the suggested selection:

```python
def show_scores(c, scores, total=False):
    suggested_select = c.autoSelect()
    score_table = scores.items()
    table_rows = ''.join("<tr%s><td>%s</td><td><b>%s</b></td></tr>" % ((' class="suggested"' if k == suggested_select else ''), categories[k], v) for k, v in score_table)
    if total:
        table_rows += "<tr><th>Total:</th><th>%s</th></tr>" % c.getTotalScore()
    return "<table><tr><th>Category</th><th>Score</th></tr>%s</table>" % table_rows

HTML(show_scores(c, c.getScoreTable()))
```

Select category and show total score:

```python
c.select(c.autoSelect())
HTML(show_scores(c, c.scores, True))
```
